```python
import os
from google.colab import userdata
```

```python
from langchain_community.llms import HuggingFaceHub

# Initialize the HuggingFace llm
llm = HuggingFaceHub(
      repo_id = "mistralai/Mistral-7B-Instruct-v0.3",
      model_kwargs = {"temperature":0.1, "max_length":500},
      huggingfacehub_api_token = userdata.get('HF_TOKEN'))
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
  model_name="sentence-transformers/all-mpnet-base-v2"
)
```

```python
from langchain.schema.output_parser import StrOutputParser
output_parser=StrOutputParser()
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("/content/test.pdf")

docs = loader.load()
```

```python
docs[0]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=50)
splits = text_splitter.split_documents(docs)
```

```python
len(splits)
```

```python
from langchain_chroma import Chroma
vectorstore = Chroma.from_documents(documents=splits, embedding=embedding_model)
```

```python
retriever = vectorstore.as_retriever()
```

```python
from langchain.prompts import ChatPromptTemplate

# Define prompt template
template = """
Answer this question using the provided context only.

# {question}

Context:
{context}

Answer:
"""

prompt=ChatPromptTemplate.from_template(template)
```

```python
prompt
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain import PromptTemplate,  LLMChain
chain = (
    {"context": retriever,  "question": RunnablePassthrough()}
    | prompt
    | llm
    | output_parser
)
```

```python
response = chain.invoke("What is Model Generalization")
answer_start = response.find("Answer:") + len("Answer:")  # Find where "Answer:" starts
answer = response[answer_start:].strip()  # Extract everything after "Answer:" and remove extra spaces
print(answer)
```

```python
import pandas as pd
from typing import List, Dict
import os

class ExcelQAProcessor:
    def __init__(self, chain):
        """
        Initialize the Excel QA processor with the existing RAG chain

        Args:
            chain: The existing RAG chain for question answering
        """
        self.chain = chain

    def process_excel(self, input_path: str, output_path: str = None) -> pd.DataFrame:
        """
        Process questions from Excel file and write answers back

        Args:
            input_path: Path to input Excel file with questions
            output_path: Optional path for output Excel file. If None, will modify input file

        Returns:
            DataFrame with questions and answers
        """

        try:
            df = pd.read_excel(input_path)
        except Exception as e:
            raise Exception(f"Error reading Excel file: {str(e)}")

        if 'Question' not in df.columns:
            raise ValueError("Excel file must contain a 'Question' column")

        if 'Answer' not in df.columns:
            df['Answer'] = ''

        for idx, row in df.iterrows():
            if pd.isna(row['Answer']) or row['Answer'] == '':  # Only process if no answer exists
                try:
                    response = self.chain.invoke(row['Question'])
                    answer_start = response.find("Answer:") + len("Answer:")
                    answer = response[answer_start:].strip()
                    df.at[idx, 'Answer'] = answer
                except Exception as e:
                    df.at[idx, 'Answer'] = f"Error processing question: {str(e)}"

        # Save results
        output_path = output_path or input_path
        df.to_excel(output_path, index=False)

        return df

    def process_questions(self, questions: List[str]) -> Dict[str, str]:
        """
        Process a list of questions and return answers

        Args:
            questions: List of questions to process

        Returns:
            Dictionary mapping questions to answers
        """
        results = {}
        for question in questions:
            try:
                response = self.chain.invoke(question)
                answer_start = response.find("Answer:") + len("Answer:")
                answer = response[answer_start:].strip()
                results[question] = answer
            except Exception as e:
                results[question] = f"Error processing question: {str(e)}"
        return results


processor = ExcelQAProcessor(chain)
input_excel = "/content/questions.xlsx"
output_excel = "/content/questions.xlsx"

try:
    results_df = processor.process_excel(input_excel, output_excel)
    print(f"Processed {len(results_df)} questions and saved to {output_excel}")
except Exception as e:
    print(f"Error processing Excel file: {str(e)}")

questions = [
    "What is Model Generalization?",
    "Another question from your PDF..."
]
results = processor.process_questions(questions)
for question, answer in results.items():
    print(f"\nQ: {question}")
    print(f"A: {answer}")
```
